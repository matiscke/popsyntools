```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tablefile = '/home/schlecker/phd/Documents/planetswMeasuredEcc.csv'
table = pd.read_csv(tablefile, sep=',', header=362, skip_blank_lines=False)
```

## Find Planets with M_P > 3e-3 M_star:

```python
def Mjup2Msol(Mjup):
    """Transfer from mass in [Jupiter mass] to mass in [Solar mass].
    """
    return 9.5458e-4*Mjup
```

```python
heavyP = table[Mjup2Msol(table['pl_bmassj']) > 3e-3*table['st_mass']]
```

```python
heavyP.pl_bmassj.hist(bins=50)
```

```python
ax = heavyP.plot.scatter('pl_orbsmax','pl_orbeccen', logx=True)
ax.set_xlabel('Orbit Semi-Major Axis [AU]')
ax.set_ylabel('Eccentricity')
plt.title('Planets with M_P > 3e-3 M_star')
```

```python
fig = ax.get_figure()
fig.savefig('Mp>3e-3Mstar_EccSma.pdf')
```
